### Notes for today 


**Polymorphism**

One name many states --> obj comes in many form 

Comes in two types:
- **Method overriding** --> child class overwrites method 
    - ```python
        class Employee:
            def display(self):
                print("This is employee")
        
        class Programmer(Employee):
            # Method overwritting 
            def display(self):
                # To use the parent's display method
                super.display() 
                print("This is programmer")
    ```
- **Method overloading** --> a method working in many different ways
    - ```python
        class Calculator:
            # Default 
            def calc(self, a=None, b=None):
                if(a is not None and b is not None):
                    print(a+b)
                elif (a is not None and b is None):
                    print(a**2)
                else:
                    print("No Data")
        
        # Method Overloading 
        c = Calculator() 
        c.calc() 
        c.calc(10,20)
        c.calc(30)
    ```

**Abstraction** 

Concerned about the larger picture (how it works are hidden to the nake eye)
    - ABC class to create an abstract class 
    - abstractmethod decortators which applys more features to current fucntion

Abstract classes are just **blueprints**
- so when you use the `@abstractmethod` decorator it can't any anythign within that function  

```python
# abc module we have an ABC class
from abc import ABC, abstractmethod

# Abstract class - More of like a Blue Print 
class Parent(ABC):
    # Decorator added to a function to apply more features 
    @abstractmethod 
    def show(self):
        # You cannot have anything within the body it's just an abstract class 
        pass

    def demo(self):
        print("Parent Class")

class Child(Parent):
    def show(self):
        print()
    

abstract_class_obj = Parent()   # Error because you cannot create an obj with abs class 

c = Child() 
c.show() 
```

### Summary

**Polymorphism**

Objects can come in **many forms**. It's a **one name - many states** 

Accomplished in two ways:
- **Method Overriding** where the child class has a method that overrides it's parent's method: *same* name *different* functionality 
- **Method Overloading** where a method has multiple functionality. It could be used in many different ways yeilding different inputs 

---

**Abstraction** 

Hiding how things work in the backend. We're more concerned with **it works** rather than **how does it work** 

So we create **blueprints** with *Abstract Class* and *Abstract Method*
- Abstract class with `from abc import ABC` and our abstract class *inherits* from the ABC class in the abc module: `class Parent(ABC)`
- Abstract methods are created with `@abstractmethod` decorator. But these classes have **no functionality**. Again it's just a blueprint.
