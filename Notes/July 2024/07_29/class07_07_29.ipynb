**Encapsulation** 

Changing memebers in the class  with `__`

```python
class Addition:
    def __init__(self):
        # Private variable 
        self.__a = 10

    # Private function
    def __changeData(self,a):
        self.__a = a 
        return self.__a

    # Public Function 
    def displayData(self,number):
        print(self.__changeData(number))
    
    # To access our private variable 
    def show_priv(self):
        print(self.__a)

add = Addition()
print(add.__a)  # Error
print(add.show_priv())  # Works 
print(add.__changeData(50)) # Error
```

**Inheritiance** 

Relationship of parent and child class by inheritance. 
- **Simple** inheritance 1 parent 1 child.
    - ```python
        class Employee:
            def __init__(self):
                self.salary = 34000
        
        # Inheritance 
        class Programmer(Employee):
            def give_bonus(self):
                # Because we inherited Employee class we have access to the attributes 
                self.salary = self.salary + 1000
                return self.salary

        # When we work with inheritance we make objects based on child classes 
        p = Programmer() 
        print(p.give_bonus())
        print(p.salary)
    ```
- **Multiple** Inheritance 
    - ```python
        class Animal:
            def eat(self):
                print("Eating")
            
        
        class Dog(Animal):
            def bark(self):
                print("Barking")
            
        class Cat(animal):
            def meow(self):
                print("Meow")
        ```
    - ```python
        class Vehicle:
            def show(self):
                print("Vehicle")
        
        class Bus:
            def display(self):
                print("Bus")
        
        # Multiple inheritance
        class Volvo(Vehicle, Bus):
            def demo(self):
                print("Volvo Vechicle Bus")
        
        v = Volvo()
        v.show()
        v.display()
        v.demo()
    ```
- **Multilevel** inheritance 
    - ```python
        class Animal:
            def eat(self):
                print("Eating")
            
        
        class Dog(Animal):
            def bark(self):
                print("Barking")
            
        # Indirectly inheriting Animal
        class BabyDog(Dog):
            def weep(self):
                print("Weeping")

    ```

````python
# Write a program using classes and objects to take the employee name and salary as input and calculate the total salary after deducting 0.2% salary from it.

class Employee:

    # Constructor 
    def __init__(self, name, salary):
        self.name = name 
        self.salary = salary 

    # Deducting 0.2% salary 
    def deduct_salary(self):
        self.salary = self.salary - (self.salary*(0.2/100))
    

e_object = Employee('Justin', 5000)
e_object.deduct_salary()
e_object.salary
````

# Summary

**Classes** 

To start with a class you need a **constructor** which is the `__init__(self, paramter)` function 

Afterwards, you could create **class methods** by creating *functions* within that class beginning the parameter with the `self` keyword => `def demo(self):`

----

**Encapsulation** 

Just means we're mangaging the members within the class. We manage with **private** vs **public** *variables* and *functions*.

We create **private** functions and variables by putting `__` in front of our declaration: 
1) `self.__priv_var`
1) `def __priv_function(self):`

Note that these private functions and variables **CANNOT** be accessed outside of the class.
- We use **public** functions to *use* and *display* private content 


---

**Inheritance**

Classes have a multitude of relationships.
1) **Simple** -> Parent:Child -> 1:1/Many (Direct inhertiance to Parent)
1) **Multiple** -> Parent:Child ->  Many:1 (Inhertining Multiple Parents)
1) **Multi-level** --> Parent:Child -> 1:2:4 (Indirectly inhertinging Parent)

With *Inheritance* comes the `super()` keyword 
- When a child inherits from a parent:
    - It **DOES NOT** run the parent's *constructor* unless the child's constructor *isn't present*
    - We have to manually run their constructor in two ways:
        1) `super.__init__(any_params)`
        1) `ClassName.__init__(self, any_params)`
    - We use ClassName if we have **multiple** or **multi-level** inheritance
