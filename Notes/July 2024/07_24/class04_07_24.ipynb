### Strings 
Every character in the string has an index. Strings are **immutable** 
- you cannot change / alter strings 

Strings are **iterable** 
```python
s = 'Hello123"
for i in s:
    print(i)

# Loops 8 times
for i in range(len(s)):
    print(s[i])
```

Strings **slicing**

```python
s = 'Hello123"
# starts at index 1 to 4 because we're stopping on 5
print(s[1:5])
# starts at index 0 going stopping on 5 
print(s[0:5])
# starts at index 0 then skips 2 which is l, then skips two which is 0
print(s[::2])
```

String **operations**
s1 = "hello123"
s2 = "world"

# Contactions 
s1 + s2 
s1 * 3

# Functions 
s = 'hello'

````python
# Input a string and count the number of letters, digits, and special characters in a string 

given_string = "testing1 23!@#"
digits = 0
letters = 0
special_char = 0

for i in given_string:
    if i.isdigit():
        digits += 1 
    elif i.isalpha():
        letters += 1
    elif not i.isalnum():
        special_char += 1

print(digits)
print(letters)
print(special_char)
````

## Summary 

Some things about Python:
1) **Strings**
    - **Immutable** cannot change / alter 
    - **Iterable** can loop through 
    - **Slicing** pick out certain parts 
    - **Operations** functions, contactions and multiply 
    - **Built-in functions**
        - `.captialize()` - like title 
        - `.upper()` - upper case
        - `.lower()` - lower case 
        - `.count(char)` - counting that character 
        - `"*".join(iterable)` - join all elements within iterable with charcter 
        - Conditional functions 
            - `.isupper()`
            - `.islower()`
            -  `isdigit()`
            - `isalpha()`
            - `isalnum()`
            - `isspace()`
1) **Lists**
    - **Container for any type** 
    - **Mutable** meaning we could alter/change 
        - Create a list from inputs 
            - use the `eval(input())` to get the correct type from inputs 
        - Loopable 
        - Slicing 
        - Operations (Concat & Multiply)
        - List functions 
            - `append()` and `extend([])` - to add one or multiple elements to the list respectively 
            - `del lst[index]` - removing an element given an index
            - `pop()` - removing and returning the final element
            - `.sort(reverse=True)` - Sorting elements (Asc - Desc | Desc - Asc)
            - `.reverse()` - reverse the list
