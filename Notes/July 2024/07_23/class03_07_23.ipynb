# Summary 

**CDN (Content Delivery Network)**
- Group of **distributed servers** that deliver internet *content* quickly to users (reduce *latency* and improve speed of content delivery)
    1) **Distribute** content to various servers. When user requests the CDN delivers it from the cloests user 
    1) **Caching** copies of content (images, video, HTML pages, and other files) on multiple servers to reduce load.
    1) ATM Analogy --> **withdraw money** from nearest ATM, a user could **access content** from the nearest CDN server.

---

**OS (Origin Server)**
- Serves as a **primary** server that **process and reponds** to incoming internet requests from clients 
    - Stores the *"Main version"* of the content and delivers to **CDN servers** for *caching and distribution*


---

**Edge Server**
- Servers **a part of the CDN** that are placed in various locations around the world 
    - **reduces the load** on OS by serving cached content to users. 
    - retrieves updated content from **OS** to cache

---

**DNS (Domain Name System)**
- Translate **domain names to IP address** for computers to identify on the network 
    - Used to access "Human-Readable websites" to "IP Addresses" (soo we don't have to remember the IP addresses)
    - How does it work?
        1) **User requests**: user starts by *typing in a domain name* (Google.com)
        1) **DNS Resolver**: this request sent to DNS resolver and 
        starts to *find ip* associated with the domain name
        1) **Root Server**: contacts root DNS which *directs to TLD (Top level Domain)* (.com, .id, .vn, .net)
        1) **TLD Server**: points resolver to authoriative DNS server for the specific TLD 
        1) **DNS Server**: contains actual IP address mapping for the domain and sends it to the resolver 
        1) **Returns IP**: DNS Resolver returns the IP address
    - How does it work in your own words...
        -DNS converts *readable domain names to IP addresses*. When a user enters a domain name, a **resolver (helper)** asks the **root server**, which then directs it to the **Top Level Domain (TLD) server** for the specific domain (e.g., .com, .net). The resolver then contacts the **DNS server** which contains the actual IP address mapping for the domain. Finally, the **resolver returns the IP address** to the user's computer.

## 10 Second look:
- **CDN (Content Delivery Network)** - group of dedicated servers used to serve and delivery your content as fast as possible 
- **OS (Origin Server)** - primary server that sits, listens and processes User request. Typically sends cachable data to Edge Servers and CDN 
- **Edge Servers** - part of CDN that is strategically placed globally to store cache data to serve to the public reducing loads on the OS 
- **DNS (Domain name service)** - converts human readable domains to Ip addresses using a resolver that asks the root server which redirect it to TLD to find the associated domain. It then asks the DNS server for IP Mapping and then returns the IP address to the browser.
- **Request and Reponse Cycle** -> Client sends a request to the server which serves a response (source code | HTML, CSS, JS) to the client. It uses an *HTTP Protocol* to send a response given headers. With the rendering engine, the client is able to view their request.
