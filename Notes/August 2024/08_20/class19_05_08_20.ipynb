3 types of Display in css:
- inline (all elements stay in line (span, a, img))
- block (begin on new line )
- inline-block (way to display)

```html

<html>
    <head>
        <style>
            .a{
                display:block;
            }
            .p{
                display:inline
            }
        </style>
    </head>
    <body>
        <!-- Both inline elements -->
        <a href='#'>Click</a>
        <img src='#'>
        <p>Block</p>
    </body>
</html>


```

**Position**

5 Values:
- Static -> by default no affect on left, right, top or bottom
- Relative -> affect on left right top or bottom
- Absolute -> Works as parent child. where child could move independently 
- Fixed -> always stuck on the page following the schrollbar 
- Sticky -> changines position based on scroll

```html

<html>
    <head>
        <style>
            .div{
                border: 2px solid black;
                /* no Effect */
                /* position: static; */
                position: relative;
                top: 10;
                left: 20;
            }
            .child{
                position:absolute;
            }
        </style>
    </head>
    <body>
        <div class='parent'>
            <div class='child'>
            </div>
        </div>
    </body>
</html>


```

### Blocks

Blocks are used to display content differently. We have three ways of displaying content:
1) **Inline** which makes all elements stay in the same line like (span, a, images)
2) **Block** which makes an element enter a differnt block (in a new line)
3) **Inline-Block** which allows *width* and *height* to take into effect

```html

<html>
    <head>
        <style>
            .a{
                display:block;
            }
        </style>
    </head>
    <body>
        <!-- Usually anchor tags are inline but we change that-->
        <a href='#'>Click here</a>
        <a href='#'>Click here 2</a>
    </body>
</html>

```
---

### CSS Position 

CSS Position allows us to *change* or *move* around certain elements. So if we set the element's position to `relative` we could start moving the *top*, *bottom*, *left*, *right* of that element

```html

<html>
    <head>
        <style>
            #my_container{
                display:block;
                border: 1px solid black;
                position: relative;
                /* You could change the distance between top, bottom, left, right */
                top: 15px;
                left: 50px;
            }
        </style>
    </head>
    <body>
        <div id='my_container'>

        </div>
    </body>
</html>

```

However, we also have 4 other positions:
1) **Static** --> Default value, no affects on top left bottom or right
1) **Absolute** --> Child could move independently doesnt have to move with the parent
1) **Fixed** --> Stuck on the page following the scrollbar
1) **Sticky** --> Changes based on scroll

### CSS Float

Floats are used to display content along side each other or in a differnet manner using **Left** and **Right**

By **floating an image to the left** and **floating the paragraph tag to it's right** we could have the image on the left and some text on the right

```html

<html>
    <head>
        <style>
            img{
                float:left;
            }
            p{
                float:right;
            }
        </style>
    </head>
    <body>
        <div id='my_container'>
            <img src="#">
            <p>My Text to the right of a picture</p>
        </div>
    </body>
</html>

```

### CSS UL

Unordered List usually have bullet points. We could change that type of style given a `list-style-type` to the *li* tags 

```html

<html>
    <head>
        <style>
            ul li{
                list-style-type: none;
                list-style-position: inside;
                display: inline-block;
            }

        </style>
    </head>
    <body>
        <ul>
            <li>Thy</li>
            <li>Justin</li>
        </ul>

    </body>
</html>

```

### CSS Forms

Forms are important to take in data from our users. We could style our input fields.

```html

<html>
    <head>
        <style>
            /* We could access a certain type of input using [] */
            input[type='submit']{
                background-color: white;
                padding: 10px;
                font-size: 20px;
            }
            /* The rest of our inputs could be different too */
            input{
                padding: 20px;
                width: 80%;
                border-radius: 5px;
            }

        </style>
    </head>
    <body>
        <form>
            <input type='text'>
            <input type='password'>
            <input type='submit' value='Submit'>
        </form>
    </body>
</html>

```
