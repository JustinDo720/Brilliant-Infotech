# Notes for today

3 types of Display in css:
- inline (all elements stay in line (span, a, img))
- block (begin on new line )
- inline-block (way to display)

```html

<html>
    <head>
        <style>
            .a{
                display:block;
            }
            .p{
                display:inline
            }
        </style>
    </head>
    <body>
        <!-- Both inline elements -->
        <a href='#'>Click</a>
        <img src='#'>
        <p>Block</p>
    </body>
</html>


```

**Position**

5 Values:
- Static -> by default no affect on left, right, top or bottom
- Relative -> affect on left right top or bottom
- Absolute -> Works as parent child. where child could move independently 
- Fixed -> always stuck on the page following the schrollbar 
- Sticky -> changines position based on scroll

```html

<html>
    <head>
        <style>
            .div{
                border: 2px solid black;
                /* no Effect */
                /* position: static; */
                position: relative;
                top: 10;
                left: 20;
            }
            .child{
                position:absolute;
            }
        </style>
    </head>
    <body>
        <div class='parent'>
            <div class='child'>
            </div>
        </div>
    </body>
</html>


```

---

**Float**

Have contents aside of each other

```html

<html>
    <head>
        <style>
            img{
                float: left
            }
            /* unless you want p to act as a block */
            p{
                float: left
            }
        </style>
    </head>
    <body>
        <img src='#'>
        <p> text </p>
    </body>
</html>

```

---

Display inline will not have an afffect with it comes to width and height 

Display block will move elements to a new line and width and height does have an effect

Display inline-block will be on the same line but there will be an effect on width and height
