**Subquery**

Query inside another larger query. (**Nested Query**) in SELECT, FROM, WHERE keywords. 
- Use comparison operators to attach subequery (inner query executes first) result passed to our query 

```sql

select col
from table1
where expression operator (select col2 from table2)

```

Display employee name with max salary

```sql

-- inner query 
select max(Salary)
from employee

-- Subquery (Max salary is this and then you use the column)
select emp_name from employee
where salary = (select max(Salary) from employee)

```

Display second highest salary from the employee table 

```sql

select emp_name 
from employee 
where salary =
    (select max(salary) from employee where salary<
        (select max(Salary) from employee));

```

Write query display all dep names where number of emp are less than 2

```sql

alter table department 
add constraint dep_fk
foreign key(EmpID)
references employee(id)

update department 
set EmpID=2
where DeptNo=1

select dept_id, dep_name, count(*)
from department 
left join employee
group by dep_name 
having count(*) < 2


```
