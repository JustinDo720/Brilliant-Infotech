## Forms

Forms are used to *submit* and *gather* data from our users to perform a certain action.

### Form Tag Submit Button

We create a form using the `<form>` tag that has an `action` property and a **submit** button to *process our form*. There's also a **reset** button to clear our form.

```html

<html>
    <body>
        <form action='http://www.google.com'>
            ....
            <input type='submit' value='Submit Form'>
            <input type='reset' value='Clear Form'>
        </form>
    </body>
</html>


```

---

### Different Input Fields 

Within our form, we have different types of **fields** but each field should have a **label**. Within these **fields** you could pass *attributes* like `type`, `name`, `value` which specifies the certain **input field**, a **variable** and (if applicable) a certain **value** respectively.

You also have additional attributes like:
- `required` which makes the input field required before submitting 
- `placeholder` which adds a placeholder text within the field 

**Text and Password Field**

Used to gather **small** texts and password from users.

**Example:**

<html>
    <body>
        <form action='http://www.google.com'>
            <label>Name</label>
            <input type='text' name='first_name' placeholder='Johnny' required>
            <label>Password</label>
            <input type='password' name='pass' required>
            <input type='submit' value='Submit Form'>
            <input type='reset' value='Clear Form'>
        </form>
    </body>
</html>

**Code**
```html

<html>
    <body>
        <form action='http://www.google.com'>
            <!-- New Code -->
            <!-- ------------------------------------------------------------------ -->
            <label>Name</label>
            <input type='text' name='first_name' placeholder='Johnny' required>
            <label>Password</label>
            <input type='password' name='pass' required>
            <!-- ------------------------------------------------------------------ -->
            <input type='submit' value='Submit Form'>
            <input type='reset' value='Clear Form'>
        </form>
    </body>
</html>
```

## CSS

Cascading Style Sheet to *style* many or specific *elements* using different properties.

### Different Ways to Implement CSS

1) **Interal CSS**
    - ```html
        <html>
            <head>
                <style>
                    p{
                        color:aqua;
                    }
                </style>
            </head>
        </html>
    ```
1) **Inline CSS**
     - ```html
        <html>
            <body>
                <p style='color:aqua '>Aqua Text</p>
            </body>
        </html>
    ```
1) **Exteneral Css** --> seperate [external css file](external_css.css)
     - ```html
        <html>
            <head>
                <link rel='stylesheet' ref='path/to/cssfile.css'>
            </head>
        </html>
    ```

### Color and Background Styling

Colors and our Background could be style using CSS. We just need to mention the `background` and also `color` 

**NOTE** we could use **id** to color a **specific** element so we apply the `id` property to our **element** then access that **element** in our styling tag using the `#element_id_name` and style it

**Example:**

<html>
    <head>
        <style>
            #my_para1 {
                color:blue;
            }
        </style>
    </head>
    <body>
        <p id='my_para1'>My colored Text</p>
    </body>
</html>

```html

<html>
    <head>
        <style>
            p {
                color:blue;
            }
            body{
                background-color: biege;
            }
        </style>
    </head>
    <body>
        <p >My colored Text</p>
    </body>
</html>

```

### Border Styling and Spacing 

Border is the **outline** in our *container*. We could change things like *color*, *width*, and a specific *style*.

As for *spacing* we have **margin** and **padding** control to give our elements some *space*.
- Margin is used for **around** the element 
- Padding is used **within** the element 

Both *margin* and *padding* has positions: **left, right, bottom, left**; however, you could just mention without the position to set a *default* size for all positions.

Instead of *id* we could just use *class*. we could provide a class name then access it in our *style* tag using `.class_name`. 

**Example:**
<html>
    <head>
        <style>
            .bordered_text {
                /* Short hand: size style color */
                border: 2px double red;
                /* Longer version */
                border-size: 2px;
                border-style: double;
                border-color: red; 
                /* Padding for inside the border */
                padding:30px;
                /* Margin for around the border */
                margin: 30px;
            }
        </style>
    </head>
    <body>
        <p class='bordered_text'>My bordered text </p>
    </body>
</html>

**Code:**
```html

<html>
    <head>
        <style>
            .bordered_text {
                 border: 2px double red;
                /* Longer version */
                border-size: 2px;
                border-style: double;
                border-color: red; 
                /* Padding for inside the border */
                padding:30px;
                /* Margin for around the border */
                margin: 30px;
            }
        </style>
    </head>
    <body>
        <p class='bordered_text'>My bordered text </p>
    </body>
</html>

```

### Links and Tables with Properties 

Links have special actions we could style like:
1) `a:link` Normal State
1) `a:visited` User already Went to the Site 
1) `a:hover` Mouse over
1) `a:active` Clicked Effect

But they **MUST** be specified in this order

**Example**
<html>
    <head>
        <style>
            .adv_link:link {
                color:aqua;
            }
            .adv_link:visited {
                color:green;
            }
            .adv_link:hover {
                color:blue;
            }
            .adv_link:active {
                color:red;
            }
        </style>
    </head>
    <body>
        <a class='adv_link'>My advanced link </a>
    </body>
</html>



**Code:**
```html

<html>
    <head>
        <style>
            .adv_link:link {
                color:aqua;
            }
            .adv_link:visited {
                color:green;
            }
            .adv_link:hover {
                color:blue;
            }
            .adv_link:active {
                color:red;
            }
        </style>
    </head>
    <body>
        <a class='adv_link'>My advanced link </a>
    </body>
</html>

```
