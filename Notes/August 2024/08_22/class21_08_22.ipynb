### Bootstrap Cards

Cards are like containers that have different sections in them. Theyre used to display information in a specific manner for instances: images, titles, paragraphs, footers, links etc

```html

<html>
    <head>
        <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/css/bootstrap.min.css" integrity="sha384-xOolHFLEh07PJGoPkLv1IbcEPTNtaed2xpHsD9ESMhqIYd0nLMwNLD69Npy4HI+N" crossorigin="anonymous">
        <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body> 
        <div class='card'>
            <img class='card-img-top' src='#' style='height:300px; width:300px;'>
            <div class='card-header'>
                Header
            </div>
            <div class='card-body'>
                <div class='card-title'>Some Title</div>
                <div class='card-text'>Some text</div>
            </div>
            <div class='card-footer'>
                Footer
            </div>
        </div>
    </body>
</html>

```

### Bootstrap Dropdown

Dropdowns we spoke earlier about, they consist of a button with a `dropdown-toggle` class with an attribute called `data-bs-toggle='dropdown'`

This triggers a class `class='dropdown-menu'`  which has items with `class='dropdown-item'`

```html

<html>
    <head>
        <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/css/bootstrap.min.css" integrity="sha384-xOolHFLEh07PJGoPkLv1IbcEPTNtaed2xpHsD9ESMhqIYd0nLMwNLD69Npy4HI+N" crossorigin="anonymous">
        <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body> 
        <button type='button' class='btn dropdown-toggle' data-bs-toggle='dropdown'>Dropdown</button>
        <div class='dropdown-menu'>
            <a href='#' class='dropdown-item'>
            <hr class='dropdown-divider'>
        </div>
    </body>
</html>

---

### Bootstrap Navbar

Navbars are **navigation bars** on the top which allows you to traverse the page.

```html

<html>
    <head>
        <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/css/bootstrap.min.css" integrity="sha384-xOolHFLEh07PJGoPkLv1IbcEPTNtaed2xpHsD9ESMhqIYd0nLMwNLD69Npy4HI+N" crossorigin="anonymous">

        <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body> 
        <!-- Centers all nav content verticle -->
        <ul class='nav justify-content-center flex-column'>
            <li class='nav-item'>
                <a href='#' class='nav-link'> Link1 </a>
            </li>
            <li class='nav-item'>
                <a href='#' class='nav-link'> Link2 </a>
            </li>
        </ul>
    </body>
</html>

```

### Bootstrap Forms

Bootstrap have their down forms that we could use. These forms are already styled

Each of these input fields must have a **check** like: `class='form-control'`, `class='form-check-input'`

```html

<html>
    <head>
       <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/css/bootstrap.min.css" integrity="sha384-xOolHFLEh07PJGoPkLv1IbcEPTNtaed2xpHsD9ESMhqIYd0nLMwNLD69Npy4HI+N" crossorigin="anonymous">

        <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body> 
        <!-- Bootstrap forms -->
         <form>
            <!-- Basic Email/Text Field -->
            <div>
                <label class='form-label'>Email: </label>
                <input type='email' class='form-control'>
            </div>
            <!-- Checkboxes -->
            <div>
                <label class='form-check-label'>
                    <input class='form-check-input' type="checkbox"> Remember Me
                </label>
            </div>
            <!-- Text Area -->
            <div>
                <textarea class='form-control' rows='5' cols='30'>
            </div>
            <!-- Select Field -->
            <div>
                <select class='form-select'>
                    <option>1</option>
                    <option>2</option>
                    <option>3</option>
                    <option>4</option>
                </select>
            </div>
            <!-- Radio Field -->
            <div>
                <input type='radio' class='form-check-input' name='gender'>
                <input type='radio' class='form-check-input' name='gender'>
                <input type='radio' class='form-check-input' name='gender'>
            </div>
        </form>
    </body>
</html>

```
