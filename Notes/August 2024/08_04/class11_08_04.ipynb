# Summary 

## Tkinter GUI 

Is a **graphical user interface** module where you could create **components/widgets** and place them within a *window*. There are some **varaibles** that help with obtaining values from these *widgets* and we specifically obtian these values through *pressing a button* which runs a *function*.

--- 

### Componenets 

These components/widgets are part of tkinter to create specific things like Text to read, Buttons to click, Entering in Data. They're just used to **interact** with the window 

1) **Label** which contains some type of **text** 
    - `tk.Label(root, text='Label text')`
1) **Entry** which is like a **form field** to collect data 
    - `tk.Entry(root)`
1) **Button** which **submits** or **runs** a function upon press-down 
    - `tk.Button(root, text='Button text', command=function)`
1) **Checkbutton** allows for **multiple selection** and works with an *integer variable*
    - `tk.Checkbutton(root, text='Option 1', offvalue=0, onvalue=1, variable=tk.IntVar())`
1) **Radiobutton** allows for **one selection** among many choices and works with an *integer variable*
    - `tk.Radiobutton(root, text="One specific choice", value=returned_integer, variable=IntVar())`
    - One thing to note is that the variables must be the **same** they must share the **same** variable 

---

### Placement 

Once these components exist, they **must** be placed within the *window* in **various** ways.

1) **Place** which uses **x,y** coordinates where (0,0) is the **top left**
    - `widget.place(x=20, y=40)`
1) **Pack** which uses **side** argument to put the widget on *top, bottom, left, right*
    - `widget.pack(side='top')`
    - it's important to note that their all **lower-case**
1) **Grid** which uses **rows and columns** 
    - `widget.grid(row=0, column=0)`

### Configurations & Extra Modules

**Configurations**

There are some basic stuff to configure like **colors** and even **changing** *entry* values.

We could set additional *arguments* like:
1) **Background** color --> `root.configure(background='orange')`
    - using the `configure()` function on our **root**
1) **Widget** coloring
    - `fg=color` -> used for *text* color
    - `bg=color` -> used for *background* color 
    - `font(18,)` -> used for *font* (here we're changing the size)
        - it's important to note that we have `(18, )` with a **trialing comma** because it's a tuple


**Extra Modules**

**Message Boxes** are used to *display* messages on a *different* window.
1) We need to create our **root window** 
1) import in our messagebox 
    - `from tkinter import messagebox`
1) use `showinfo` to display your message 
    - `messagebox.showinfo('Title', 'Message')`
