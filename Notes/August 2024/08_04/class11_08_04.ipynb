## Notes for class today 


### Tkinter GUI 

**Graphical User Interface** 

You could create *componenents*:
- **label** --> `Label(root, text='text')` description / text on screen 
- **Entry** --> `Entry(root)` Put data in 
- **Button** --> `Button(root, text='text', commond=function)` Click  function used with `entry.get()`
- **Checkbutton** --> `Checkbutton(root, text='Text', offvalue=0, onvalue=1, variable=check1)` used with *variables* so `check1=tkinter.IntVar()`
    - then just use `check1.get()` 
- **Radiobutton** --> `Radiobutton(root, text='text', value=int, variable=IntVar())` Used to select one value among a lot of values 


You could place these *componenets* on the window:
- *place* - used with x,y cords (0,0) is top left
- *pack* - used with side where side could be left, right, top, bottom 
- *grid* - rows and columns 

You Have *variables* to link with *componenets*:
- `IntVar()` - Integer variable
- `StrVar()` - String variable 

Other modules:
- **messagebox** - `messagebox.showinfo("Title", "Message")` to show a message on another window box 
- **preview**


Configurations:
- **Background color** -> `root.configure(background='orange')`
    - `fg=white` - text color 
    - `bg=black` - background color 
    - `font(18, )` - font size (18, ) --> single value in tuple must need comma

````python
import tkinter 

# Creating Root 
root = tkinter.Tk() 

# Two input fields with labels 
num_1 = tkinter.Entry(root)
operation_lb = tkinter.Label(root)
num_2 = tkinter.Entry(root)
equal_lb = tkinter.Label(root, text='=')
ans_display = tkinter.Label(root)

num_1.pack(side='left')
operation_lb.pack(side='right')
num_2.pack(side='right')
equal_lb.pack(side='right')
ans_display.pack(side='right')


root.mainloop()
````
