### ORDER BY 

Orders a specific column in **ASC** order; however, you could change that by adding in a keyword **DESC**

```sql

select * from table_name
order by col;

select * form table_name
order by col DESC;

```

### GROUP BY and HAVING

Perform an *aggregate functions* onto a **shared value** within a column. We could also restrict our results using the `HAVING` keyword based on the **aggregate results**

```sql

-- count() could be sub by any aggregate function
SELECT col1, count(*)
FROM table_name
GROUP BY col1
HAVING count(*) > 5

```

What this does is that it will count all **occurance** (because we're using the **aggregate `count()` function**) of *col1*. The `HAVING` clause restricts our results to show where the **occurances** of values within *col1* are greater than 5. 

---

### Aggregate Functions

These functions perform a **mathematical operation** on a specific column.

**Count** 

Looks at the **amount present** within a column. Used with `GROUP BY` or the `WHERE` clause to obtain a specific count.

```sql

SELECT count(col1)
FROM table_name
WHERE condition

SELECT col2, count(col1)
FROM table_name
GROUP BY col2

```

**Average** 

Looks at the **central** (average) number of a set.

```sql

SELECT avg(col1)
FROM table_name

SELECT col2, avg(col1)
FROM table_name
GROUP BY col2
HAVING avg(col1) > 53.5

```

**Sum**

Looks at the **total** of a certain column.

```sql

SELECT sum(col1)
FROM table_name

SELECT col2, sum(col1)
FROM table_name
GROUP BY col2
HAVING sum(col1) < 50

```

**Min and Max**

Looks for the **minimum or maximum** number within a column 

```sql

SELECT min(col1)
FROM table_name
WHERE condition

SELECT col2, max(col1)
FROM table_name
GROUP BY col2

```

### Distinct Values

Grabs the unique value within a column.

```sql

ALTER TABLE table_name 
ADD COLUMN col_date DATE DEFAULT (CURDATE());


UPDATE table_name
-- 'YYYY-MM-DD'
SET col_date = '2021-10-30'
WHERE ID = primary_int

INSERT INTO table_name(col_date)
VALUES ('2021-10-30')

-- Distinct Values
SELECT DISTINCT(col_date)
FROM table_name;

```
