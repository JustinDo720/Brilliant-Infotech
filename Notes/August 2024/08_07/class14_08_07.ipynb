**Having**

Groups by will use HAVING

```sql

SELECT col
FROM table_name
WHERE condition 
GROUP BY column_name
HAVING condition 
ORDER BY condition

```

"List number of employees in each city  that has more than 2 employee"

```sql

SELECT country, count(*)
FROM customer
WHERE country is not null
GROUP by country
HAVING count(*) > 2;

```

Here HAVING deals with **aggregate** function (condition on aggregate) you cannot apply WHERE because it doesn't checck aggregate 

Country isnt UK and the count of country is greater than 2
```sql

SELECT country, count(*)
FROM customer
WHERE country <> 'uk'
GROUP by country
HAVING count(*) > 2;

```
