### JavaScript Variables

We start **Javascript** by providing a `<script>` tag in our **body** or **reference** it in our **header**. This **client-side** language is a scripting language in which we could use **variables**

There are 3 types of ways to declare a variable:
1) `let var = 'value'` where the *let* keyword **doesn't let** you access variables outside the block 
1) `var var_name= 'value'` where the *var* keyword **let you** access variables outside the block 
1) `const const_name = 'constant value'` where the *const* keyword doesn't let you **change** the value 

```html

<html>
    <head>
        <script src='path_to_our_js_file.js'></script>
    </head>
    <body>
        <!-- Or we could just write JS in our script tag in the body section -->
        <script>
            const msg = 'My Message'

            document.write(msg)
        </script>
    </body>
</html>


```

### JavaScript Data Types, Operators

Javascript has the same **data types** as Python: Numbers, Booleans, Floats, Strings

It also has the same operators: +, -, /, *, %, **

And same comparison operators: !=, ==, <=, >=, <, >

But we have new operators like: 
1) `&&` for **logically and** 
1) `||` for **logically or**
1) `!()` for **logically not** 

We also have something called **Ternary Operators** which is like a **one-liner** 
`var result = condition?"result_true_value":"result_false_value"`

So this would have an output of `"A is less than 10"`:
```js

let a = 5 
let result = a<10:'A is less than 10':'A is Greater than 10'

```

**Increments** and **Decrements** are also different in JS 
- **Post-Increment** --> `a++` --> Increments the **code after** 
- **Pre-Decrements** --> `--a` --> Decrements **immediately**

So depending on **where the sign is** it might increment or decrement **after** or **immediately**

### JavaScript Functions and Array Functions

We could create functions which are **reusable** piece of code. But we have different ways to write functions:
1) **Normal** way with the *function* keyword 
1) **Anonymous** Functions which sets a *variable* to to our function
1) **Arrow** Functions which uses an `=>` to create our functions 


```js 

// Normal 
function function_name(a,b=20){
    document.write(a+b)
    return a+b
}

function_name(1)

// Anonymous 
ana = function anonymous_function(p1, p2){
    document.write(p1,p2)
}

ana(1,2)

// Arrow 

arr = (para1, para2) => {
    console.log(para1, para2)
}

arr(20,30)

```

**Array Functions** exist to perform iterations on it much **faster**

We use the syntax of:

```js

forEach(function(value, index, array){
    // some execution 
})

```

Like:


```js

let arr=[2,5,10,20]
arr.forEach(function(value, index, array){
    document.write(value**2+"<br>")
})

```
