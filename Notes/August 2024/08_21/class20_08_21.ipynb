### Font Awesome

We could search for icons to put into our website by going to [**Font Awwesome**](https://fontawesome.com/icons) and looking for their icons which provide a code similar to `<i class="fa-solid fa-heart"></i>`; however, we need to provide a font awesome **CDN**: `https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.6.0/css/all.min.css`

With this library we could use their icons after creating a *link* tag in our *header* and then grabbing the `<i>` from their website


```html

<html>
    <head>
        <link rel='stylesheet' href='https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.6.0/css/all.min.css'>
        <style>
            .fa-soli{
                color:red;
            }
        </style>
    </head>
    <body>
        <a style='text-decoration: none'> No Line Link </a>
        <i class="fa-solid fa-heart"></i>
    </body>
</html>

```

### Bootstrap Table

Boostrap Tables are quite **important** but we have the same structure as an **HTML** table structure:
- `<table>` which has:
    - `<tr>` for a **table row** and `<th>` for a **table header**
    - `<tr>` for a **table row** and `<td>` for a **table data**

But instead we also have **two parts:**
- `<thead>`
- `<tbody>`

```html

<html>
    <head>
        <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.0.2/dist/css/bootstrap.min.css" rel="stylesheet" integrity="sha384-EVSTQN3/azprG1Anm3QDgpJLIm9Nao0Yz1ztcQTwFspd3yD65VohhpuuCOmLASjC" crossorigin="anonymous">
        <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body> 
        <div class='container'>
            <table class='table table-striped table-bordered table-hover table-info'>
                <!-- Table head which contains our headers -->
                <thead>
                    <!-- Row full of <th> which are table headers -->
                    <tr>
                        <th> Name </th>
                        <th> Age </th>
                    </tr>
                </thead>
                <!-- Table Body which includes our data -->
                <tbody>
                    <!-- Table Row full of <td> which are table data -->
                    <tr>
                        <td>
                            Name1
                        </td>
                        <td>
                            Age1
                        </td>
                    </tr>
                </tbody>
            </table>
        </div>
    </body>
</html>

```

### Bootstrap Buttons

Using some Bootstrap preset classes, we could style our buttons. But just remember that all our buttons must:
1) Have a `type='button'`
1) Have a `class='btn'`

With **dropdown** buttons we must use a class of `dropdown-toggle` and a `data-bs-toggle='dropdown'` attribute 
- After creating a button we need to make a `div` which has a class of `dropdown-menu` 
- Then inside that div we have elements with class `dropdown-item`

```html

<html>
    <head>
        <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.0.2/dist/css/bootstrap.min.css" rel="stylesheet" integrity="sha384-EVSTQN3/azprG1Anm3QDgpJLIm9Nao0Yz1ztcQTwFspd3yD65VohhpuuCOmLASjC" crossorigin="anonymous">
        <meta name="viewport" content="width=device-width, initial-scale=1">
    </head>
    <body> 
        <div class='container'>
            <button type='button' class='btn btn-outline-secondary btn-lg'>Click</button>
            <!-- We could also make a dropdown button -->
            <button type='buton' class='btn dropdown-toggle' data-bs-toggle='dropdown'>
                Button name
            </button>
            
            <!-- Now for we dropdown menu which opens whenever we click that button -->
            <div class='dropdown-menu'>
                <a class='dropdown-item' href='#'>Table</a>
                <a class='dropdown-item' href='#'>Row</a>
                <a class='dropdown-item' href='#'>columns</a>
            </div>

        </div>
    </body>
</html>

```
