### Hashmaps

Hashmaps are basically **python dictionaries**. But in python keys are strings, here keys are **not strings**

```js

// We use this. to access the example hashmap WITHIN the example hashmap 

var example = {
    first_name = 'Thy',
    last_name = 'Nguyen',
    fullname = ()=>{
        return this.first_name + ' ' + this.last_name
    }
}

example.fullname()
```

### String Functions

We have JS string functions to transform strings to *uppercase* and *lowercase* or *combine* two strings.

We could also find the length (similar to the `len()` function) using the `.length`

```js

var word = 'Hello'

// Upper and Lower case 
console.log(word.toUpperCase())
console.log(word.toLowerCase())

// Combining
console.log(word.concat(word, word))

// Length 
console.log(word.length)

```

### DOM

The **Document Object Model** is constructed like a **tree** which consist of our **HTML**. This helps us create that **JS Window** 
and let us run actions like:

```js

let name = prompt('Question ')
let conf = confirm('Confirming')

```

But It also allows us to **find** elements via *element id* or *tag name* 

```html

<html>
    <body>
        Name:
        <input type='text' id='name'>
        <script>
            function grab_name(){
                // Working with the DOM
                alert(document.getElementById('name').value)
            }
        </script>
    </body>
</html>

```

Cool, we created the function. However, how will our input know when to **trigger** that function. Well we have **events for that**
---

### JS Events

Have certain Types of **Events** like:
1) `onclick()`
1) `mouseoverevent()`
1) `onfocus()`
1) `onchange()`
1) `onkeyup()`

So we add these **events** to our elements to **trigger** our function 

```html

<html>
    <body>
        Name:
        <!-- NEW CODE HERE -->
        <!-- -------------------------------------------------------------------------------- -->
        <input type='text' id='name' onchange='grab_name()'>
        <!-- -------------------------------------------------------------------------------- -->
        <script>
            function grab_name(){
                // Working with the DOM
                alert(document.getElementById('name').value)
            }
        </script>
    </body>
</html>

```
