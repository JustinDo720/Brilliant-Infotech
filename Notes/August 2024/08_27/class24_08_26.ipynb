**Callbacks** waiting for a function to run first 

```js

// Callback function runs after a another function is finished 
// Function waits for another function
function greet(name){
    console.log(name)
}

function show(func){
    func('John')
}

show(greet)

```

---

**Promise** object that have multiple states that lets us work after some code as been execute.
- Pending --> undf result
- Resolved --> fulfilled result
- Rejected --> error


```js

var promise = new Promise((resolve, reject)=>{
    var x=10, y=20
    if(x==y){
        resolve()
    } else {
        reject()
    }
})

promise.then(()=>{
    console.log('Promise Resolved.')
}).catch(()=>{
    console.log('Promise Rejected. We will handle errors')
})
```

### Async

Javascript is asyncronous meaning it **doesn't wait** for each line to execute.

But to understand this we need to understand **callback functions** which as we said it's just using a function within another function 

```js

function show_name(name){
    console.log(name)
}

// using show_name as a callback function 
function call_name(func){
    func('Thy')
}

call_name(show_name)

```

---

### Awaits

Awaits lets you wait for a certain **request** to start before continuing the code. We just need to use the `await` keyword 

Usally you use this with *axios* and make an **API call**

```js

// specify that our function is going to be async to use the await keyword 
async function display(){
    // awaits for this line of code to execute 
    await let msg = 'hello'
    console.log(msg)
}

```

You usually use **async** with **API calls** because it returns a *promise*

```js

const axios = require('axios');

async function fetchData(){
    // waits for this axios.get() request to finsih before executing more code
    const rep = await axios.get('https://jsonplaceholder.typicode.com/posts/1')
    // because of await console.log() will execute whenever the GET request finishes 
    console.log(rep.data)
}

```

### Promises

Has three states:
1) **Pending** where the result is undefined 
1) **Resolved** when the result is fulfilled 
1) **Rejected** when the result returns an error 

You could *manually* create a Promise 

```js

var promise = new Promise((resolve, reject)=>{
    if(condition){
        resolve()
    } else {
        reject()
    }
})

// Here we could check the result if it's resolved or rejected
// If resolved then it will run the .then() block
// If rejected then it will run the .catch() block
promise.then(()=>{
    console.log('Resolved')
}).catch(()=>{
    console.log("Rejected")
})

```
