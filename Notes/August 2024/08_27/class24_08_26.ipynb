**Callbacks** waiting for a function to run first 

```js

// Callback function runs after a another function is finished 
// Function waits for another function
function greet(name){
    console.log(name)
}

function show(func){
    func('John')
}

show(greet)

```

---

**Promise** object that have multiple states that lets us work after some code as been execute.
- Pending --> undf result
- Resolved --> fulfilled result
- Rejected --> error


```js

var promise = new Promise((resolve, reject)=>{
    var x=10, y=20
    if(x==y){
        resolve()
    } else {
        reject()
    }
})

promise.then(()=>{
    console.log('Promise Resolved.')
}).catch(()=>{
    console.log('Promise Rejected. We will handle errors')
})
```

```js
// returns a promise soo we need to resolve our promise
async function functionName(){
    return "Hello"
}
// call the function then use .then()
functionName().then((value)=>{
    console.log(value)
}).catch(()=>{
    console.log('err')
})

// Await can we used inside the async function
async function display(){
    await let msg = 'hello'
    console.log(msg)
}

const display2= async()=>{
    await let data='Hello World';
    return data
}

console.log(1)
console.log(display2())
console.log(2)

let promise = new Promise((resolve, reject)=>{
    setTimeout(()=>{
        resolve('Promise Resolved')
    }, 40000)
})

async function func({
    // await waits for your promise it will wait until the promise is fullfilled before running the rest of the codce
    let result = await promise;
    console.log(result)
    console.log("Hello")
})

// promise chaining 
let promise = new Promise((resolve, reject)=>{
    setTimeout(()=>{
        resolve(1)
    }, 1000)
}).then((result)=>{
    alert(result)
    return result * 2
}).then((result)=>{
    return result * 4
})
```

### Async

Javascript is asyncronous meaning it **doesn't wait** for each line to execute.

But to understand this we need to understand **callback functions** which as we said it's just using a function within another function 

```js

function show_name(name){
    console.log(name)
}

// using show_name as a callback function 
function call_name(func){
    func('Thy')
}

call_name(show_name)

```
