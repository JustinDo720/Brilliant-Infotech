## Starting Html

To begin HTML you need to have that **template** which consists of:
1) Head --> containing meta information like scripts 
1) Body --> Holds your designing elements 

We start with the `<html>` **tag**

```html

<html>
    <head>
        ....
    </head>
    <body>
        ....
    </body>
</html>

```

---

## Text Elements

Taking a look at **elements** which are present within the `<head>` and `<body>` tags
1) `<title>` works as a meta information which presents a title for your **tab**
1) `<hx>` where *x* is an *int* to resize your **heading** like: `<h1>`
1) `<p>` is a **paragraph** tag which allows you to write text
1) `<u>` is an **underline** tag which underlines a certain text
1) `<b>` **boldens** your selected word(s)
1) `<mark>` **highlights** your selected word(s)
1) `<sub>` converts your content to a **subscript**
1) `<sup>` converts your content to a **superscript**


**Example:**
<html>
    <head>
        <title>My Webpage</title>
    </head>
    <body>
        <h3>Heading 3<h3>
        <p>My Paragraph tag</p>
        <br>
        <u>Underlined text</u>
        <br>
        <b>Bolded Work</b>
        <br>
        <mark>Highing my text</mark>
        <br>
        Subscript<sub>word</sub>
        <br>
        Superscript<sup>word</sup>
        <br>
        <abbr title='Dunkin Donuts'> DD (Abbreivations)</abbr>
    </body>
</html>

Some other elements like `<abbr>`, `<img>` and `<a>` requires some **attributes** in order for the *abbreivation*, *image*, and *link* to work

## Tables 

We could create **tables** using html to provide *data* in **rows and columns**. We use four major **tags**
1) `<table>` to begin our **table**
1) `<tr>` to begin a **row** 
1) `<th>` for the **header** row
1) `<td>` to provide a **data** row

```html

<html>
    <head>
        <title>My Webpage</title>
    </head>
    <body>
        <!-- Starting the table with borders-->
        <table border='2'>
            <!-- First row of headers-->
            <tr>
                <!-- Specifiying Header -->
                <th>First Header</th>
                <th>Second Header</th>
            </tr>
            <!-- First row of Data-->
             <tr>
                <!-- Specifiying Data -->
                <td>First Data</td>
                <td>Second Data</td>
            </tr>
        </table>
    </body>
</html>
```

We could also customize with `colspan` and `rowspan` to take up more space or *span* across a *row* or *column*

## Ordered and Unordered List

**Ordered List** is just **numbered** or have **alphabets** depending on the `type` attribute 
- `type='a'` lowercase alphabet (if `A` it will be uppercase)
- `type='i'` lowercase romain numeral (if `I` it will be uppercase)

```html

<!-- Starting ordered List -->
<ol>
    <li>First List Item</li>
    <li>Second List Item</li>
    <li>Third List Item</li>
</ol>

<ol type="A">
    <li>First List Capital Alphabet Item</li>
    <li>Second List Capital Alphabet Item</li>
    <li>Third List Capital Alphabet Item</li>
</ol>

```

**Unordered List** works the same way except it's in **bullet format**

```html

<!-- Starting unordered List -->
<ul>
    <li>First Unordered List Item</li>
    <li>Second Unordered List Item</li>
    <li>Third Unordered List Item</li>
</ul>

<ul type="square">
    <li>First List squared Item</li>
    <li>Second List squared Item</li>
    <li>Third List squared Item</li>
</ul>

```

**Description List** is similar to a *keyword* and *definition* pair 

```html

<!-- Description List -->
<dl>
    <!-- Description Term -->
    <dt>
        HTML Term
    </dt>
    <!-- Description Term -->
    <dd>
        Description for HTML term
    </dd>

</dl>

```
