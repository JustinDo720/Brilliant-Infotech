# Notes for Today

**HTML** -> frontend code to interact with the browser 
- create a file with `.html` extension
    - Start with a `<html>` tag 
    - `<head>` will contian meta information like title scripts etc
    - `<body>`
- Tags 
    - `<p>`
    - `<h1>`
    - `<br>`
    - `<hr>`
    - `<i>` `<b>` `<mark>`  `<u>`
    -
    



```html

<html>
    <head>
        <!-- Meta information -->
        <title>
            My Title 
        </title>
    </head>
        <!-- Actual Display elements on the web-->
    <body bgcolor='black'>
        <p>
            Paragraph tag
        </p>
        <h1>
            Heading 1 
        </h1>
         <h2>
            Heading 2
        </h2>
         <h3>
            Heading 3 
        </h3>

        <!-- Empty Tag to break a line  -->
        <p>
            Some text first line <b> <br> Next line within the paragrpah tag <b>
        </p>

        <!-- Drawing a line -->
        <hr>
        <u>
            Underline text
        </u>
        <i>
            Italizied text
        </i>

        <!-- highlighting -->
        <p>
            <mark>
                Some Text
            </mark>
        </p>

        <!-- subscript -->
         <p>
            H
            <sub>
                2
            </sub>
            O
        </p>
        <!-- superscript -->
         <p>
            A
            <sup>
                3
            </sup>
        </p>
        
        <!-- Meaning to acronym -->

         <abbr title="World Human Organization">W.H.O</abbr>

         <!-- Align -->
        <p align='right'>
            <font  size="20" color="red">
                Text
            </font>
        <p>
        
        <!-- Html Links  and opening link to new tab -->
        <a href='link' target='_blank'>
            click here 
        </a>

        <!-- Image Tag -->
         <img src='path/to/image' width=750 height=750>
    </body>

</html>



```

```html
<html>
    <head>
        <title>
            My Website
        </title>
    </head>
    <body>
        <a href="https://www.amazon.com/">
            <img src="amazon-3932136766.jpg" width="250" height="250">
        </a>
    </body>
</html>
```

**Tables** 

```html

<table border="2">
    <!-- Row 1 -->
    <tr>
        <th>
            Column1 Name 1
        <th>
        <th>
            Column1 Name 2
        <th>
    </tr>

    <!-- Row 2 -->
    <tr>
        <td>
            Col1 data
        </td>
        <td>
            Col2 data
        </td>
    </tr>
    
    <!-- Row 3 -->
    <tr>
        <td>
            Col1 data
        </td>
        <td>
            Col2 data
        </td>
    </tr>

</table>

<!-- One column having two data -->
<table border="2">
    <!-- Row 1 -->
    <tr>
        <th>
            Column1 Name 1
        <th>
        <th colspan='2'>
            Column1 Name 2
        <th>
    </tr>

    <!-- Row 2 -->
    <tr>
        <td>
            Col1 data
        </td>
        <td>
            Col2 data
        </td>
        <td>
            Col2 Columnspan Data
        </td>
    </tr>
</table>
```

**List**

**Ordered List** has numbers or alphabets 
- Depending on the `type` by default its numbers 
    - Alphabet `type=a` or capital 
    - Romain Numeral `type=i` or capital

```html

<ol type='A'>
    <li>
        Tea
    </li>
    <li>
        Coffee
    </li>
    <li>
        Milk 
    </li>
</ol>

```

**Unordered List**
 
 Bullet formats

```html

<ul type='square'>
    <li>
        Tea
    </li>
    <li>
        Coffee
    </li>
    <li>
        Milk 
    </li>
</ul>

```
