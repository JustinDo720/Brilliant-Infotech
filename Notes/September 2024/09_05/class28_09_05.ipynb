**Conditions**

If blocks 

Filters:
- `|` 

So something like: `{% if num|divisibleby:"2" %}`

Filters:
- | capfirst - Upper() 
- | add:'2' - Adding 2 to that number
- | center:20 - Centers a word in a block (try textarea )
- | date:"D d M y" --> use a date object and send it to the template with this filter 
- | dictsort:'key'
- | length 
- | upper
- | lower
- | wordcount 
- | slice:"1:5" --> exluding 5 
- | random to grab a random valu from a list 

```html

<html>
    <body>
        {% if name %}
            {{ name | capfirst}}
        {% elseif %}
            Nolo
        {% endif %}
        
        {% for i in num %}
            {% if i|divisibleby:"2"}
                {{ i | add:'2'}}
            {% endif %}
        {% endfor %}

        {% now "d\t\h\o\f\ M Y }
    </body>
</html>

```

Include is used to Load another template 

`{% include "template_name.html" %}`

Similar kind of design so you dont want to design it again and again if it keeps appearing 

Like making a *navbar* or a *footer* 

Send variables to templates 

*menu.html* has two variables soo we could use `with` to declare the variable 
`{% include 'menu.html' with me='name' sponser='me'}`

You could also make a variable with Django template iwth `with`

```html

{% with name='Thy'}
    {{ name }}
{% endwith %}

```
