If blocks

- `|`

- | capfirst - Upper() 
- | add:'2' - Adding 2 to that number
- | center:20 - Centers a word in a block (try textarea )
- | date:"D d M y" --> use a date object and send it to the template with this filter 
- | dictsort:'key'
- | length 
- | upper
- | lower
- | wordcount 
- | slice:"1:5" --> exluding 5 
- | random to grab a random valu from a list

## Jinja For Loops

We know **Jinja** tags look something like this: `{{ }}`` however, we could also loop through a given variables passed during the **render phase** that are *arrays*

We just use the `{% for var in vars %}` but we always need to **end it**: `{% endfor %}`

```html

<html>
    <body>
        {% for i in names %}
            {{ i }}
        {% endfor %}
    </body>
</html>


```

---

## Jinja Conditionals

We have if blocks in Jinja with the syntax of: `{% if var %}` and `{% elseif %}` and as always we need to **end it** `{% endif %}`

```html
<html>
    <body>
        {% if name %}
            {{ name | capfirst}}
        {% elseif %}
            Nolo
        {% endif %}
    </body>
</html>

```

Now Here you saw `{{ name | capfirst}}` where the `|` is actually a filter and there's a bunch of them like:
- | length 
- | upper
- | lower

---

## Jinja Including

Say we have a **reocurring** template code like: a *navbar* or a *footer* which needs to be present throughout all templates...

We could write a navbar and footer then include them:
`{% include "template_name.html" %}`

Sometimes we have variables for these templates so we could pass variables using `with`:
`{% include 'menu.html' with me='name' sponser='me'}`

We could also use `{% with var='value' %}` and `{% endwith %}` in a regular template
