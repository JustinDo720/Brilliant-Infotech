# Model Forms 

Saves form and saves into Model (Convert Model into Django Form)

1) Create your Model first with all the fields
1) create forms.py and inherit from forms.ModelForm
- ```python
    class StudentForm(forms.ModelForm):
    class Meta:
        model = StudentModel
        fields = '__all__'
    ```

then we import it into our **views** and add to it the **context** which is passed to our template


```python

def demo_form(request):
    student = StudentForm()
    context = {}
    context['student'] = student 
    return render(request, 'demo_form.html', context)

```

Now within our **template** we need a form tag and a **csrf_token**

```html

<html>
    <body>
        <!-- If using file must use enctype -->
        <form method="post" enctype="multipart/form-data">
            {% csrf_token %}
            {{
                student
            }}
            <input value="submit"/>
        </form>
        
    </body>
</html>

```

**Template inheritance**

Common sections and reuse in other templates 

```html

<html>
    <head>
        <title>{% block title %} My current page's title {% endblock %} </title>
    </head>
    <body>
        {% block body %}
        {% endblock %}
    </body>
</html>

```

You could inherit this base.html to another 

```html

{% extends 'base.html' %}
<html>
    <head>
        {% block title %} My New Title {% endblock %}
    </head>
    <body>
        {% block body %}
        <H1>Welcome to my new page </H1>
        {% endblock %}
    </body>
</html>

```

Here all styles within base.html will be kept but the content is different so 
1) style are same content will be different
    - Title is different 
    - Body is didferent
