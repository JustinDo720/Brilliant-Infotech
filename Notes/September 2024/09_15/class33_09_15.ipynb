**Many-to-Many** Fix 

```python

class Publication(models.Model):
    title = models.CharField(max_length=50)

class Article(models.Model):
    headline = models.CharField(max_length=80)
    # Publication object 
    publication = models.ManyToManyField(Publication)


```

Then we go ahead and `py manage.py makemigrations` and then we migrate so: `py manage.py migrate` 

Before adding the **publication** object you must make the *Article* object first then **save it** 

`a1 = Article(headline='Hello')`

Then save it 

`a1.save()`

Then We could go ahead and use the **add** function 

`a1.publication.add(Publication)` where `Publication` is actually a publication object 

In **Many-To-Many** it will create **three** tables.

**Class based views**

Class based views create **crud** operations becuase they are already built in

let's make a new app and register i in settings:
`py manage.py startapp new_app_name` 

So let's make a models and make a **Class Based View** 

*views.py*

```py 

from django.views.generic.edit import CreateView
from .models import * 

class EmployeeCreate(CreateView):
    model = EmployeeModel
    fields='__all__'


```

This will **automatically** search for your own **template** but it must follow the pattern:

*app_name/template/app_name/my_new_template.html*

Now that we created the **ModelForm** with the *Meta* class of our model and full fields 

We need to to tie it in the **urls** 

`from AppName.views import ClassBasedViewName` 

```python

urlpatterns = [
    path('create_view_path/', ClassBasedViewname.as_view())
]

```
