# Foreign Keys

Foreign keys are used to **link** Two models. This allows models to hav relationships with each other which means there are going to be different ways in which we use *foreign keys*

---

# One-To-Many And Many-To-One

Here is where we have a *Parent* and *Child* class where One model could have many instances of another like:

One **Person** eating many **Food**

*app_name.models.py*

```python

class Person(models.Model):
    ...

class Food(models.Model):
    ...
    # then a reference to Person 
    person = models.ForeignKey(Person, on_delete=models.CASCADE)

```

We provide that ForeignKey with the reference **class** and what happens when that class is deleted

We could **add** data using **django shells**:

`py manage.py shell`

**Creating a person instance**

`p = Person(field=value, field2=value2)`

Then we save:

`p.save()`

We could also create one using `p = Person.objects.create(field=value, field2=value2)`

Let's make a **Food** instance:

`f = Food(field=value, field2=value2, person=p)`
`f.save()`

Remember that *p* is that **Person instance** or person object 

Now There will be a link between that food with a Person 
(One-Many or Many-One)

# One-To-One

Inserting data will be the **same** but creating a One-to-One will be **different**

We create a relationship of:

One **Person** having One **Salary**

*app_name.models.py*

```python

class Person(models.Model):
    ...

class Salary(models.Model):
    ...
    # then a reference to Person 
    person = models.OneToOneField(Person, on_delete=models.CASCADE)

```

# Many-To-Many

Is Different in the sense of **adding** and **creating** the relationship 

Let's say a **Project** has many **Students** but students could also have many projects

We need to create three classes

*app_name.models.py*

```python

class Student(models.Model):
    ...

class Project(models.Model):
    ...
    # then a reference to Person 
    Student = models.ManyToMany(Student, through='StudentProject')


class StudentProject(models.Model):
    student = models.ForeignKey(Student, on_delete=models.CASCADE)
    project = models.ForeignKey(Project, on_delete=models.CASCADE)

```
