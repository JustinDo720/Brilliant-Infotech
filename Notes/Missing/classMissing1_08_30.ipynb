## Array as Prop 

Let say we passed an array as a Prop

```js

const root = ReactDOM.createRoot(document.getElementById('root'));
root.render(
  <React.StrictMode>
    <Demo arr=[12m, 13m, 17m]></Demo>
  </React.StrictMode>
);

```

We passed an array to a *Demo Componenet* so how do we display **arrays**

We just use the `props.arr_name.map()` function in which it cycles through all the items within the array and display them on a HTML template:

`map((item)=>{html})`

```js

function Demo(props){
    return(
        <>
            {props.arr.map((item)=>{
                <li>{item}</li>
            })}
        </>
    )
}

```

# Creating Refs

We could create references for events. On a certain event we could change aspects of the elements and we access those with **references**


```js

function Demo(props){

    // Creating that Ref for events
    const inputRef = React.createRef()
    const second_inputRef = React.createRef()

    // Function to change our background 
    const changeBackground= () =>{
        inputRef.current.style.backgroundColor='Blue'
    }

    const alert_name = () =>{
        alert(second_inputRef.current.value)
    }

    return(
        <>
            <input type='text' ref={ inputRef } onFocus={changeBackground}/>
            <input type='text' ref = {second_inputRef }>
            <input type='button' onClick='alert_name'>
        </>
    )
}

```

# States 

These are used like a variable but we could update them throughout 

You could `useState()` and set some default values. but remember to set **two varaibles**
- state_name
- functionToUpdateState

```js

import { useState } from "react"
function Demo(props){

    const [data, setData] = useState({
        name:'Thy',
        salary: 75000
    })

    const changeValue = () =>{
        setData(data.salary = 80000)
    }

    return(
        <>
            {{ data }}

            <button onClick={changeValue}>
        </>
    )
}

```
