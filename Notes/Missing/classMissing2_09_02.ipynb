# Parent and Child

We could create child components and **render the parent**

*child.js*
```js

function Child(props){
    return(
        <>
            {{prop.message}}
        </>
    )
}

export default Child;

```

now in *parent.js*
```js
// import child component
import Child from './child'

function Parent(){
    return(
        <>
            <Child message='my msg'>
        </>
    )
}


```
